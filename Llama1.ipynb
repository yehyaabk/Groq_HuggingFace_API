```python
import transformers
import torch

model_id = "meta-llama/Meta-Llama-3.1-8B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},
    device_map="auto",
)
```

```python
messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "Who are you?"},
]

outputs = pipeline(
    messages,
    max_new_tokens=256,
)
print(outputs[0]["generated_text"][-1])
```

```python
messages = [
    {"role": "system", "content": "You are a pirate chatbot who always responds in pirate speak!"},
    {"role": "user", "content": "what is python language"},
]

outputs = pipeline(
    messages,
    max_new_tokens=256,
)
print(outputs[0]["generated_text"][-1]['content'])
```

```python
def run_chat(pipeline, user_prompt, system_prompt="You are a helpful assistant.", max_new_tokens=256):
  messages=[
      {'role':'system', 'content':system_prompt},
      {'role':'user', 'content':user_prompt}
  ]

  outputs = pipeline(
      messages,
      max_new_tokens=max_new_tokens
  )

  return outputs[0]["generated_text"][-1]["content"]
```

```python
print(run_chat(pipeline, "what is the capital of switzerland"))
```

```python
print(run_chat(pipeline, "what is the difference between DL and ML", "you are a second year computer science student"))
```
