```python
from groq import Groq
from IPython.display import display, Markdown
import re
import os
```

```python
client = Groq(
    api_key=os.environ.get("api_key")
)
```

```python
def ask_groq(user_prompt, system_prompt="You are a helpful assistant.", model="qwen/qwen3-32b", temperature=0.8):
    messages = [
        {'role': 'system', 'content': system_prompt},
        {'role': 'user', 'content': user_prompt}
    ]

    reply = client.chat.completions.create(
        model=model,
        temperature=temperature,
        messages=messages
    )
    return reply.choices[0].message.content
```

```python
def remove_thinking(text):
    search_pattern = r'<think>.*?</think>'
    clean_text = re.sub(search_pattern, "", text, flags=re.DOTALL)
    return clean_text.strip()
```

```python
ask_groq("what is the difference between Deep Learning and Machine Learning")
```

```python
Markdown(ask_groq("what is the difference between Deep Learning and Machine Learning"))
```

```python
Markdown(remove_thinking(ask_groq("what is the difference between Deep Learning and Machine Learning")))
```

# Model Sentence Completion

```python
Markdown(remove_thinking(ask_groq("On a cold winter morning, the city was unusually quiet...", system_prompt="you are a creative story writer")))
```

# Summarization

```python
long_text = """
In recent years, artificial intelligence has become an increasingly influential force across many industries.
Companies now rely on AI systems to analyze large volumes of data and automate repetitive tasks.
In healthcare, AI assists doctors by detecting diseases earlier.
Despite these benefits, AI also raises ethical concerns.
"""
```

```python
Markdown(remove_thinking(ask_groq(long_text, system_prompt="you are a professional summarizer")))
```
